```python
def car_num(x):
    pattern = r'^[АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2}\d{2,3}$'
    number = r'\d{3}'
    region = r'\d{2,3}$'
    y = re.findall(pattern, x)
    if len(y) != 0:
        print(int(re.findall(number, x)[0]))
        print(int(re.findall(region, x)[0]))
car_num(input('enter car number: '))
```

```python
def valid_mail(mail):
    pattern = r'^[\w\.\+\-]+\@[\w]+\.[a-z]{2,4}$'
    if re.findall(pattern, mail):
        return 'ok'
    else:
        return 'wrong mail!'
valid_mail(input('enter ur e-mail: '))
```

```python
def valid_number(number):
    pattern = r'^(\s*)?(\+)?([- _():=+]?\d[- _():=+]?){10,14}(\s*)?$'
    pattern_num = re.compile(r'\d+')
    if re.findall(pattern, number):
        return (''.join(pattern_num.findall(number)))
    # Дальше вероятно нужно провернуть что-то с re.sub()




print(valid_number(input('Введите номер телефона: ')))
```
