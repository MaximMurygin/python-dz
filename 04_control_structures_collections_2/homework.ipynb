```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
    'виноделие'
]
d = {}
for i in range(len(queries)):
    d[queries[i]] = len(queries[i].split(' '))
# print(d)
views1 = 0
views2 = 0
views3 = 0
for views in d.values():
    if views == 1:
        views1 += 1
    elif views == 2:
        views2 += 1
    elif views == 3:
        views3 += 1
total = views1 + views2 + views3

print('Из одного слова: ', (views1 / total * 100), '%')
print('Из двух слов: ', (views2 / total * 100), '%')
print('Из трех слов: ', (views3 / total * 100), '%')
```
