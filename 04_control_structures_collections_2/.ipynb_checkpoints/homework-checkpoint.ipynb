```python
from pprint import pprint
geo_logs = [
    {'visit1': ['Москва', 'Россия']},
    {'visit2': ['Дели', 'Индия']},
    {'visit3': ['Владимир', 'Россия']},
    {'visit4': ['Лиссабон', 'Португалия']},
    {'visit5': ['Париж', 'Франция']},
    {'visit6': ['Лиссабон', 'Португалия']},
    {'visit7': ['Тула', 'Россия']},
    {'visit8': ['Тула', 'Россия']},
    {'visit9': ['Курск', 'Россия']},
    {'visit10': ['Архангельск', 'Россия']}
]
geo_logs_rus = []
country = 'Россия'
for visit in geo_logs:
    if list(visit.values())[0][1] == country:
        geo_logs_rus.append(visit)
pprint(geo_logs_rus)
```

```python
queries = [
    'кулинария',
    'новости спорта',
    'смотреть фильмы онлайн',
    'афиша кино театр спектакли',
    'рецепт суп том ям видео',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
    'виноделие',
    'сериалы с субтитрами'
]
d = {}

for query in queries:
    d[(len(query.split(' ')))] = 0
for query in queries:
    for key in d.keys():
        if len(query.split(' ')) == key:
            d[key] += 1
for words,values  in d.items():
    print(f'Запросов из {words} слов: {round(values/len(queries)*100, 2)} %')
```
