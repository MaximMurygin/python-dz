```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}


def doc_name(doc_list):
    x = input('Введите номер документа:    ')
    for doc in doc_list:
        if doc['number'] == x:
            return doc['name']
        else:
            return print('Документа с таким номером не существует!')


def doc_shelf(shelfs):
    doc_num = input('Введите номер документа:')
    for shelf, docs in shelfs.items():
        if doc_num in docs:
            return shelf
        else:
            print('Документа с таким номером не существует!')


def show_list(documents):
    for i in documents:
        print(f"{i['type']} '{i['number']}' '{i['name']}'")


def add_shelf(shelfs):
    number_shelf = input('Введите номер полки: ')
    shelfs.setdefault(number_shelf, [])
    return shelfs


def del_doc(doc_list, shelfs):
    doc_number = input('Введите номер удаляемого документа: ')
    var = 0
    for i in doc_list:
        if doc_number == i['number']:
            del doc_list[var]

        var += 1

    for shelf, docs in shelfs.items():
        if doc_number in docs:
            shelfs[shelf] = [num for num in docs if num != doc_number]
            return print(documents, shelfs, sep='\n')

        else:
            return 'номер не найден!'


def remove(shelfs):
    doc_number = input('Введите номер документа: ')
    shelf_number = input('Переместить на полку: ')
    for shelf, doc in directories.items():
        if doc_number in doc:
            directories[shelf] = [num for num in doc if num != doc_number]
            directories[shelf_number].append(doc_number)
            return shelfs
            return shelfs

        elif doc_number not in doc:
            print('Неверный номер документа!')
            return shelfs
        elif shelf_number not in shelf:
            print('Данная полка не существует!')
            return shelfs


def add_doc(doc_list, shelfs):
    doc_list.append(
        {'type': input('введите тип: '), 'number': input('введите номер документа: '), 'name': input('введите имя: ')})

    shelf_number = input('На какой полке разместить? ')
    for shelf, docs in shelfs.items():
        if shelf_number == shelf:
            shelfs[shelf_number].append(doc_list[len(doc_list) - 1]['number'])
            return print(documents, shelfs, sep='\n')
        elif shelf_number not in shelfs.keys():
            print('Данная полка не существует!')
            return print(documents, shelfs, sep='\n')


def main():
    while True:
        a = input('Введите команду: ')
        if a == 'l':
            return show_list(documents)
        elif a == 's':
            return doc_shelf(directories)
        elif a == 'p':
            return doc_name(documents)
        elif a == 'as':
            return add_shelf(directories)
        elif a == 'd':
            return del_doc(documents, directories)
        elif a == 'm':
            return remove(directories)
        elif a == 'a':
            return add_doc(documents, directories)


print(main())
```
