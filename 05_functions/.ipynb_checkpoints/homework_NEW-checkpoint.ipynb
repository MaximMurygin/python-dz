```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}


def doc_name(doc_list):
    x = input('Введите номер документа: ')
    counter = 0
    for doc in doc_list:
        counter += 1
        if doc['number'] == x:
            counter += 1
            print(doc['name'])
            break
        if counter == len(doc_list):
            print('Документ не найден!')


def doc_shelf(shelfs):
    doc_num = input('введите номер документа: ')
    doc_list = [] 
    for i in shelfs.values():
        for j in i:
            doc_list.append(j)
    for shelf, doc in shelfs.items():
        if doc_num in doc:
            print(shelf)
        elif doc_num not in doc_list:
            print('Документа с таким номером не существует!')
            break  # чтобы 3 раза не выводился "else"


def show_list(documents):
    for i in documents:
        print(f"{i['type']} '{i['number']}' '{i['name']}'")


def add_shelf(shelfs):
    number_shelf = input('Введите номер полки: ')
    shelfs.setdefault(number_shelf, [])
    return shelfs


def del_doc(doc_list, shelfs):
    doc_number = input('Введите номер удаляемого документа: ')

    for doc in doc_list:
        if doc_number in doc.values():
            doc_list.remove(doc)
    for shelf, docs in shelfs.items():
        if doc_number in docs:
            shelfs[shelf] = [num for num in docs if num != doc_number]
            print(documents, shelfs, sep='\n')

            break

    else:
        print('номер не найден!')


def remove(doc_list, shelfs):
    import sys
    doc_number = input('Введите номер документа: ')
    numbers = []
    for doc in doc_list:
        for i in list(doc.values()):
            numbers.append(i)
    if doc_number not in numbers:
        print('Документ не найден!')
        sys.exit()
    shelf_number = input('Переместить на полку: ')
    if shelf_number not in shelfs.keys():
        print('Полка не существует!')
        sys.exit()
    for shelf, doc in shelfs.items():
        if doc_number in doc:
            shelfs[shelf] = [num for num in doc if num != doc_number]
            shelfs[shelf_number].append(doc_number)
    print(shelfs)


def add_doc(doc_list, shelfs):
    doc_list.append(
        {'type': input('введите тип: '), 'number': input('введите номер документа: '), 'name': input('введите имя: ')})

    shelf_number = input('На какой полке разместить? ')
    for shelf, docs in shelfs.items():
        if shelf_number == shelf:
            shelfs[shelf_number].append(doc_list[len(doc_list) - 1]['number'])
            return print(documents, shelfs, sep='\n')
        elif shelf_number not in shelfs.keys():
            print('Данная полка не существует!')
            del doc_list[-1] # удаляет созданный словарь при отсутствии подходящей полки
            return print(documents, shelfs, sep='\n')


def main():
    while True:
        a = input('Введите команду: ')
        if a == 'l':
            show_list(documents)
        elif a == 's':
            doc_shelf(directories)
        elif a == 'p':
            doc_name(documents)
        elif a == 'as':
            add_shelf(directories)
        elif a == 'd':
            del_doc(documents, directories)
        elif a == 'm':
            remove(documents, directories)
        elif a == 'a':
            add_doc(documents, directories)


print(main())
```
