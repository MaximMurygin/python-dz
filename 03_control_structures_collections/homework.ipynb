```python
a = input('введите слово: ')
if len(a) % 2 != 0:
    print(a[len(a) // 2])
else:
    print(a[len(a) // 2 - 1] + a[len(a) // 2])
```

```python
boys = input('Введите имена парней через пробел: ').split(' ')
girls = input('Введите имена девушек через пробел: ').split(' ')
boys, girls = sorted(boys), sorted(girls)
import sys

if len(boys) != len(girls):
    print('Количество людей в списках не совпадает!')
    sys.exit()
print('Идеальные пары:')
for i in range(len(boys)):
    print(boys[i] + ' и ' + girls[i])
```

```python
countries_temp = [
    ['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Germany', [57.2, 55.4, 59, 59, 53.6, 55.4, 57.2]],
    ['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4, 51.8]],
]
for i in range(len(countries_temp)):
    mid_temp = sum(countries_temp[i][1]) / len(countries_temp[i][1])
    print(countries_temp[i][0] + ': ' + str((5 / 9) * (mid_temp - 32)) + '°C')
```

```python
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]
views = 0
users = 0
for i in range(len(stream)):
    stream[i] = stream[i].split(',')
    views += int(stream[i][2])
    if stream[i][1] != stream[i - 1][1]:
        users += 1
print('В среднем просмотров на пользователя: ' + str(views/users))
```

```python
stats = [
    ['2018-01-01', 'google', 25],
    ['2018-01-01', 'yandex', 65],
    ['2018-01-01', 'market', 89],
    ['2018-01-02', 'google', 574],
    ['2018-01-02', 'yandex', 249],
    ['2018-01-02', 'market', 994],
    ['2018-01-03', 'google', 1843],
    ['2018-01-03', 'yandex', 1327],
    ['2018-01-03', 'market', 1764],
]
date = input('Введите дату: ')
company = input('Введите название компании: ')
for i in range(len(stats)):
    if date == stats[i][0] and company == stats[i][1]:
        print(stats[i][2])
```

```python
import random

x = int(input('Enter the size of matrix: '))
data = []
for i in range(x):
    data.append([])
    for j in range(x):
        data[i].append(random.randint(1, 99))

summary = 0
for i in range(len(data)):
    print(data[i])
    summary += data[i][i]
print(summary)
```
