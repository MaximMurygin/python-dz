```python
sal = int(input('Введите заработную плату в месяц: '))
mortg_percent = int(input('Сколько процентов от ЗП уходит на ипотеку: '))
exp_percent = int(input('Сколько процентов от ЗП уходит "на жизнь": '))
awarded = int(input('Сколько раз приходит премия в год: '))
total = sal * 12 + sal * awarded / 2
mortgage = sal * 12 * mortg_percent / 100
spent = sal * 12 * exp_percent / 100
collected = total - mortgage - spent
print('На ипотеку было потрачено: ', mortgage, 'рублей')
print('Было накоплено: ', collected, 'рублей')
```
