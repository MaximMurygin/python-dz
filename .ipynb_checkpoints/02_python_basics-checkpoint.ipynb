```python
month = input('Введите месяц: ')
day = int(input('Введите число: '))
if month == 'январь' and 31 >= day > 20 or month == 'февраль' and 1 <= day < 20:
    print('Водолей')
elif month == 'февраль' and 29 >= day > 19 or month == 'март' and 1 <= day < 21:
    print('рыбы')
elif month == 'март' and 31 >= day > 20 or month == 'апрель' and 1 <= day < 20:
    print('овен')
elif month == 'апрель' and 30 >= day > 20 or month == 'май' and 1 <= day < 21:
    print('телец')
elif month == 'май' and 31 >= day > 20 or month == 'июнь' and 1 <= day < 22:
    print('близнецы')
elif month == 'июнь' and 30 >= day > 21 or month == 'июль' and 1 <= day < 23:
    print('рак')
elif month == 'июль' and 31 >= day > 22 or month == 'август' and 1 <= day < 24:
    print('лев')
elif month == 'август' and 31 >= day > 21 or month == 'сентябрь' and 1 <= day < 23:
    print('дева')
elif month == 'сентябрь' and 30 >= day > 22 or month == 'октябрь' and 1 <= day < 24:
    print('весы')
elif month == 'октябрь' and 31 >= day > 23 or month == 'ноябрь' and 1 <= day < 23:
    print('скорпион')
elif month == 'ноябрь' and 30 >= day > 22 or month == 'декабрь' and 1 <= day < 22:
    print('стрелец')
elif month == 'декабрь' and 31 >= day > 21 or month == 'январь' and 1 <= day < 21:
    print('козерог')
else:
    print('wrong input')
```
